## **Series**

In this column we are looking at the Olympic game years in a descending format from most recent years to earlier. The data provided starts with the most recent year 2016 in Rio and earlier. We want to narrow down the Olympic events to the last 8 years(2008) to find the recent athlete that can represent Kellogg's. Candidates from earlier events are not in consideration due to lack of media exposure and wont contribute to Kellogg's brand recognition. 

In order to narrow down the selection within the 2008-2016 time frame, a command  "descend" was created to descend the years from recent to earliest. By using the ".loc" function, we can then narrow down the years from 2008 to 2016 while displaying the years in a descending order. By having the yeares formated in the decending order, we can clearly see the data was filter correctly with most recent 5 rows being from 2016 and last 5 rows being the earliest time in 2008.

```python
descend = data.sort_values(by = "Year", ascending = False)
```

```python
years = descend.loc[descend.Year >= 2008, ]
```

```python
years
```

This is the year of the summer Olympics so we want to select atletes from the summer games to sponsor. By choosing candidates from the summer games it will bring more attention to the audience and result to more revenue. 

After selecting the time frame with "years", we continue to filter down the selection to summer games using ".lo" and creating a new command "summer" where it displays only athletes from the summer. 

```python
summer = years.loc[years.Season == "Summer", ]
```

```python
summer
```

Now that we have the correct season and recent years of candidates, we now need to look into sponsoring the best athlete. In order to be consider the best, they must be a gold medalist winner. By selecting the best, it will help the marketing team sell the idea to the public. 

Again we will use the ".loc" function to filter out non-gold medalist winner so we a data frame of only the gold medalist winners. 

```python
summer.loc[summer.Medal == "Gold", ]
```
